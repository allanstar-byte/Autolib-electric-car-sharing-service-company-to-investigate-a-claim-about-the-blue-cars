## **7. Hypothesis testing**

Null Hypothesis: People with great freedom of making life choices and  a higher Gdp per capita will report greater happiness score than people with less freedom of making life choices and less Gdp per capita

Alternative Hypothesis:People with great freedom of making life choices and  a higher Gdp per capita will not report greater happiness score than people with less freedom of making life choices and less Gdp per capita.

### **7.2 Parametric Test**

The main reason of performing the parametric and non-parametric tests is to get a better understanding of our core parameters. That is, GDP per capita and the happiness score.


#### **Z Test**

##### **a) One-Sample Z-test**

Ho = The GDP per capita and Freedom to make choices population mean is equal GDP per capita and Freedom to make choices sample mean

H1 = The GDP per capita and Freedom to make choices population mean is not equal GDP per capita and Freedom to make choices sample mean

The test will tell us whether means of the sample and the population are different

```
#selecting the GDP per capita and freedom to make choices columns
gdp = df['GDP_per_capita']

free = df['Freedom']
```

```
# getting the mean of the GDP per capita
gdp_mean = np.mean(gdp)
gdp_mean
```

```
## getting the mean of the freedom of choices
free_mean = np.mean(free)
free_mean
```

```
#importing libraries
from scipy import stats
from statsmodels.stats import weightstats as stests

# carrying out the Z-test for the GDP per capita
ztest ,pval = stests.ztest(df['GDP_per_capita'], x2=None, value=156)
print(float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

```
# carrying out the Z-test for the freedom of choice
ztest ,pval = stests.ztest(df['Freedom'], x2=None, value=156)
print(float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

we're going to reject the null hypothesis since the p value is less than 0.05. Thus there is a significant difference between the sample and the population mean with a p-value of 0.0

##### **b) Two-sample Z-test**

The two sample Ztest compares the means of two independent groups in order to determine whether there is a statistical evidence that the associated population means are significantly different.

H0 = The GDP per capita and Freedom to make choices population mean have no statistical difference with GDP per capita and Freedom to make choices population means

H1 = The GDP per capita and Freedom to make choices population mean have a statistical difference with GDP per capita and Freedom to make choices population means

```
#carrying out the Z-test
ztest ,pval1 = stests.ztest(df['GDP_per_capita'], x2=df['Freedom'], value=0,alternative='two-sided')
print(float(pval1))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

From the results obtained, there is a statistical difference between the sample and the population means

### **7.3 non-parametric**

#### **Chi-squared goodness of fit**

H0 = A variable follows a hypothesized distribution.

H1 = A variable does not follow a hypothesized distribution.

```
#creating arrays of GDP per capita and the freedom to make choices
capita = df['GDP_per_capita']
choices = df['Freedom']
```

```
import scipy.stats as stats

#perform Chi-Square Goodness of Fit Test
stats.chisquare(f_obs=capita, f_exp=choices)
```

Since the p-value (0.0) is less than 0.05, we alternative hypothesis. This means we have sufficient evidence to say that the true distribution of GDP per capita is different from the distribution of the Freedom to make choices.
